```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Data.xlsx')
```

```python
df.columns
```

```python
#remove unwanted columns like New Actor
df.drop(['New Actor', 'New Music Director', 'Lead Star', 'Whether Remake', 'Whether Franchise', 'Music Director', 'Director', 'New Actor', 'Category', 'Movie Name', 'Release Period', 'Genre', 'New Director', 'Revenue/Screen', 'Revenue/Budget'], axis=1, inplace=True)
```

```python
df.head()
```

```python
from IPython.display import IFrame
wiki = IFrame("https://en.wikipedia.org/wiki/Pearson_correlation_coefficient", width=1000, height=600)
display(wiki)
```

```python
#Create a correlation matrix for Revenue(INR)
corr = df.corr()
```

```python
#visualize the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt
#size of the plot
sns.set(rc={'figure.figsize':(12,8)})
#plot the correlation matrix of Revenue(INR) and Budget(INR) in the same plot
sns.heatmap(corr, annot=True, cmap='viridis')

#set title
plt.title('Correlation Matrix of Bollywood Movies', fontsize=30, fontname='Trebuchet MS', pad=20)

#save the plot
plt.savefig('Correlation Matrix of Bollywood Movies.png', dpi=300, bbox_inches='tight')
```
